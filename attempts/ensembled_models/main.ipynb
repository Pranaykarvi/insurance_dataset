```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, LSTM, GRU, Conv1D, Flatten, Concatenate, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

# Assuming X is your feature matrix
# 1. Handle datetime columns
if 'Policy Start Date' in data.columns:
    data['Policy Start Date'] = pd.to_datetime(data['Policy Start Date'])
    data['Policy Start Date'] = data['Policy Start Date'].apply(lambda x: x.timestamp())

# 2. Encode categorical columns
categorical_cols = data.select_dtypes(include=['object', 'category']).columns
for col in categorical_cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])

# 3. Handle missing values
data = data.fillna(0)


```

```python
X, y = load_data(data)
```

```python
def create_ensemble_model(input_shape):
    # LSTM branch
    lstm_input = Input(shape=(input_shape[0], input_shape[1]))
    lstm_layer = LSTM(64, activation='relu', return_sequences=True)(lstm_input)
    lstm_output = Flatten()(lstm_layer)

    # GRU branch
    gru_input = Input(shape=(input_shape[0], input_shape[1]))
    gru_layer = GRU(64, activation='relu', return_sequences=True)(gru_input)
    gru_output = Flatten()(gru_layer)

    # CNN branch
    cnn_input = Input(shape=(input_shape[0], input_shape[1]))
    cnn_layer = Conv1D(64, kernel_size=3, activation='relu')(cnn_input)
    cnn_output = Flatten()(cnn_layer)

    # Concatenate outputs
    combined = Concatenate()([lstm_output, gru_output, cnn_output])
    dense_layer = Dense(128, activation='relu')(combined)
    dropout_layer = Dropout(0.3)(dense_layer)
    output = Dense(1, activation='linear')(dropout_layer)

    # Define the model
    model = Model(inputs=[lstm_input, gru_input, cnn_input], outputs=output)
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse', metrics=['mae'])

    return model
```

```python
def train_model(model, X_train, y_train, X_val, y_val, batch_size=32, epochs=50):
    # Train the model
    history = model.fit(
        [X_train, X_train, X_train],  # Inputs for each branch
        y_train,
        validation_data=([X_val, X_val, X_val], y_val),
        batch_size=batch_size,
        epochs=epochs,
        verbose=1
    )
    return history
```

```python
X_reshaped = X.reshape(X.shape[0], X.shape[1], 1)

# Split data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_reshaped, y, test_size=0.2, random_state=42)

# Create the model
input_shape = (X_train.shape[1], X_train.shape[2])
ensemble_model = create_ensemble_model(input_shape)

# Train the model
history = train_model(ensemble_model, X_train, y_train, X_val, y_val)

# Save the model
ensemble_model.save('insurance_premium_ensemble_model.h5')
```
