```python
import pandas as pd

# Load your data
# data = pd.read_csv('your_dataset.csv')  # Uncomment and specify your file path

# Helper function for filling missing values and mapping
def fill_and_map(data, column, fill_value, mapping):
    data[column] = data[column].fillna(fill_value)
    data[column] = data[column].replace(mapping)
    return data

# Gender mapping
data['Gender'] = data['Gender'].replace({'Male': 1, 'Female': 2})

# Marital Status mapping
data = fill_and_map(data, 'Marital Status', 'Unknown', {'Married': 1, 'Divorced': 0, 'Single': 2, 'Unknown': -1})

# Number of Dependents mapping
data = fill_and_map(data, 'Number of Dependents', 'Unknown', {'Unknown': -1})

# Annual Income mapping
data = fill_and_map(data, 'Annual Income', 'Unknown', {'Unknown': -1})

# Age mapping
data = fill_and_map(data, 'Age', 'Unknown', {'Unknown': -1})

# Education Level mapping
data['Education Level'] = data['Education Level'].replace({'Bachelor\'s': 1, 'High School': 2, 'Master\'s': 3, 'PhD': 0})

# Occupation mapping
data = fill_and_map(data, 'Occupation', 'Unknown', {'Employed': 1, 'Unemployed': 0, 'Self-Employed': 2, 'Unknown': -1})

# Health Score mapping
data = fill_and_map(data, 'Health Score', 'Unknown', {'Unknown': -1})

# Location mapping
data['Location'] = data['Location'].replace({'Urban': 1, 'Rural': 2, 'Suburban': 3})

# Policy Type mapping
data['Policy Type'] = data['Policy Type'].replace({'Basic': 1, 'Comprehensive': 2, 'Premium': 3})

# Previous Claims mapping
data = fill_and_map(data, 'Previous Claims', 'Unknown', {'Unknown': -1})

# Vehicle Age mapping
data = fill_and_map(data, 'Vehicle Age', 'Unknown', {'Unknown': -1})

# Credit Score mapping
data = fill_and_map(data, 'Credit Score', 'Unknown', {'Unknown': -1})

# Insurance Duration mapping
data = fill_and_map(data, 'Insurance Duration', 'Unknown', {'Unknown': -1})

# Customer Feedback mapping
data = fill_and_map(data, 'Customer Feedback', 'Unknown', {'Average': 1, 'Poor': 0, 'Good': 2, 'Unknown': -1})

# Smoking Status mapping
data['Smoking Status'] = data['Smoking Status'].replace({'Yes': 1, 'No': 0})

# Exercise Frequency mapping
data['Exercise Frequency'] = data['Exercise Frequency'].replace({'Monthly': 1, 'Weekly': 2, 'Daily': 3, 'Rarely': 0})

# Property Type mapping
data['Property Type'] = data['Property Type'].replace({'House': 1, 'Condo': 2, 'Apartment': 3})

# Display transformed data (optional)
# print(data.head())

# Export the transformed data if required
# data.to_csv('processed_data.csv', index=False)  # Uncomment to save the transformed dataset

```
